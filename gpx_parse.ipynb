```python
fname = '/content/run.gpx'
layer = fiona.open(fname, layer='tracks')
geom = layer[0]
data = {'type': 'MultiLineString','coordinates': geom['geometry']['coordinates']}
shp = shape(data)
m = folium.Map()
folium.FitBounds([[layer.bounds[1],layer.bounds[0]],[layer.bounds[3],layer.bounds[2]]]).add_to(m)
folium.GeoJson(shp).add_to(m)
m
```
