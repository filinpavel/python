```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Генерация данных
np.random.seed(42)
data = np.random.rand(100, 3)

# Применение PCA
pca = PCA(n_components=2)
transformed_data = pca.fit_transform(data)

# Визуализация
plt.figure(figsize=(8, 6))
sns.scatterplot(x=transformed_data[:, 0], y=transformed_data[:, 1])
plt.title('PCA: Снижение размерности до 2D')
plt.xlabel('Главная компонента 1')
plt.ylabel('Главная компонента 2')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Генерация данных
np.random.seed(42)
X = np.random.rand(100) * 10
Y = 2.5 * X + np.random.randn(100) * 2

# Линейная регрессия
coefficients = np.polyfit(X, Y, deg=1)
Y_pred = np.polyval(coefficients, X)

# Визуализация
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X, y=Y, label='Данные', color='blue')
plt.plot(X, Y_pred, color='red', label='Линия регрессии')
plt.title('Метод наименьших квадратов')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import lsq_linear

# Генерация данных
np.random.seed(42)
X = np.random.rand(100, 1) * 10
Y = 2.5 * X.flatten() + np.random.randn(100) * 2

# Ограниченная регрессия (коэффициенты >= 0)
result = lsq_linear(X, Y, bounds=(0, np.inf))
Y_pred = result.x * X.flatten()

# Визуализация
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X.flatten(), y=Y, label='Данные', color='blue')
plt.plot(X, Y_pred, color='red', label='Линия регрессии с ограничением')
plt.title('Ограниченная линейная регрессия')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Генерация данных
np.random.seed(42)
data = np.vstack([np.random.randn(100, 2) + [3, 3], np.random.randn(100, 2) - [3, 3]])

# Применение K-средних
kmeans = KMeans(n_clusters=2, random_state=42).fit(data)
clusters = kmeans.labels_

# Визуализация кластеров
sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=clusters, palette='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Центроиды')
plt.title('Метод K-средних')
plt.xlabel('Признак 1')
plt.ylabel('Признак 2')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

# Генерация данных
np.random.seed(42)
X_class_1 = np.random.randn(50, 2) + [2, 2]
X_class_0 = np.random.randn(50, 2) - [2, 2]
X = np.vstack([X_class_1, X_class_0])
y = np.hstack([np.ones(50), np.zeros(50)])

# Обучение модели логистической регрессии
model = LogisticRegression().fit(X, y)

# Визуализация разделения классов
xx, yy = np.meshgrid(np.linspace(-5, 5), np.linspace(-5, 5))
grid = np.c_[xx.ravel(), yy.ravel()]
probs = model.predict_proba(grid)[:, 1].reshape(xx.shape)

sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette='viridis', s=50)
plt.contour(xx, yy, probs, levels=[0.5], colors='red')
plt.title('Логистическая регрессия: Граница решений')
plt.xlabel('Признак 1')
plt.ylabel('Признак 2')
plt.show()
```

```python
from sklearn.svm import SVC

# Обучение SVM-классификатора
model_svm = SVC(kernel='linear').fit(X, y)

# Визуализация SVM-гиперплоскости
decision_function = model_svm.decision_function(grid).reshape(xx.shape)
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette='viridis', s=50)
plt.contour(xx, yy, decision_function, levels=[-1, 0, 1], linestyles=['--', '-', '--'], colors='red')
plt.title('Метод опорных векторов (SVM)')
plt.xlabel('Признак 1')
plt.ylabel('Признак 2')
plt.show()
```
